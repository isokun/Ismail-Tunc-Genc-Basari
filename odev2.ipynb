```python
import requests
from datetime import datetime, timedelta

API_KEY = "29bd73e9806c4917ad9413a8b5ac552c"
BASE_URL = "https://newsapi.org/v2/everything"

saved_news = []

# READ
def fetch_news(query="tesla", date_range="weekly"):
    today = datetime.today()

    if date_range == "daily":
        from_date = today - timedelta(days=1)
    elif date_range == "weekly":
        from_date = today - timedelta(weeks=1)
    elif date_range == "monthly":
        from_date = today - timedelta(days=30)
    else:
        from_date = today - timedelta(weeks=1)

    url = f"{BASE_URL}?q={query}&from={from_date.strftime('%Y-%m-%d')}&sortBy=publishedAt&apiKey={API_KEY}"

    response = requests.get(url)
    data = response.json()

    return data.get("articles", [])

# CRUD
class news_manager:
    @staticmethod
    def create_news(news):
        saved_news.append(news)
        print("\n‚úÖ Haber ba≈üarƒ±yla kaydedildi.")

    @staticmethod
    def read_news():
        if not saved_news:
            print("\n‚ö†Ô∏è Kaydedilmi≈ü haber bulunmuyor.")
        else:
            print("\nüìå Kaydedilen Haberler:")
            for idx, news in enumerate(saved_news, start=1):
                print(f"{idx}. {news}")

    @staticmethod
    def update_news(index, new_content):
        if 0 <= index < len(saved_news):
            saved_news[index] = new_content
            print("\n‚úÖ Haber ba≈üarƒ±yla g√ºncellendi.")
        else:
            print("\n‚ö†Ô∏è Ge√ßersiz indeks, haber bulunamadƒ±.")

    @staticmethod
    def delete_news(index):
        if 0 <= index < len(saved_news):
            deleted = saved_news.pop(index)
            print(f"\nüóëÔ∏è '{deleted}' haberi silindi.")
        else:
            print("\n‚ö†Ô∏è Ge√ßersiz indeks, haber bulunamadƒ±.")

# Tarihler
def user_input_fetch():
    print("\nüîç Hangi zaman dilimindeki haberleri g√∂rmek istersiniz?")
    print("1 - G√ºnl√ºk\n2 - Haftalƒ±k\n3 - Aylƒ±k")
    choice = input("Se√ßiminizi girin (1/2/3): ")

    if choice == "1":
        return "daily"
    elif choice == "2":
        return "weekly"
    elif choice == "3":
        return "monthly"


# Save
selected_range = user_input_fetch()
news_list = fetch_news(date_range=selected_range)

print("\nüîπ G√ºncel Haberler:")
for idx, news in enumerate(news_list[:5], start=1):
    print(f"{idx}. {news['title']} - {news['source']['name']}")

# Save
if news_list:
    news_manager.create_news(news_list[0]["title"])

# ƒ∞lk haberi silme
news_manager.delete_news(0)

# Son haliyle listeleme
news_manager.read_news()
```
